```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bankMar_db = pd.read_csv('bank-additional-full.csv')
```

```python
# Feature variables
X = bankMar_db[['age', 'duration']]

# Target Variables
y = bankMar_db['y']
```

# Spliting and building the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.40,random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=90, metric='euclidean').fit(X_train, y_train)
```

```python
# Accuracy
model.score(X_train, y_train)
```

```python
# Test accuracy
model.score(X_test, y_test)
```

```python
train_accuracy = []
test_accuracy =[]

kvals = [i for i in (1, 100)]

for k in kvals:
    model = KNeighborsClassifier(n_neighbors=k, metric='euclidean').fit(X_train, y_train)
    train_accuracy.append(model.score(X_train, y_train))
    test_accuracy.append(model.score(X_test, y_test))
    
plt.plot(kvals, train_accuracy)
plt.plot(kvals, test_accuracy)
```
